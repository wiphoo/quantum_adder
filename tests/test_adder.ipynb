```python
from adder import Adder
```

```python
num_bits = 2
adder = Adder()
```

```python
a = '01'
b = '11'
adder.draw_initialize_input_circuit(a, 'latex')
```

```python
adder.draw_initialize_input_circuit(b, 'latex')
```

```python
adder.draw_qft_circuit(num_bits, 'latex')
```

```python
adder.draw_add_circuit(num_bits, 'latex')
```

```python
adder.draw_inverse_qft_circuit(num_bits, 'latex')
```

```python
adder.draw_adder_circuit(a, b, 'latex')
```

```python
(quantum_circuit, output_register) = adder.create_adder_circuit(a, b)
quantum_circuit += adder.create_measure_circuit(output_register)
```

```python
quantum_circuit.draw('latex')
```

# run an experiment on IBM Q

```python
import qiskit
```

```python
backend = qiskit.BasicAer.get_backend( 'qasm_simulator' )
result = qiskit.execute( quantum_circuit, backend ).result()
print(f'a={a}, b={b}')
print(f'result={result.get_counts()}')
```
